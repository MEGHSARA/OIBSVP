```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris_df = sns.load_dataset('iris')
```

```python
print(iris_df.head())
```

```python
print(iris_df.describe())
```

```python
sns.pairplot(iris_df, hue='species')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for i, feature in enumerate(iris_df.columns[:-1]):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(x='species', y=feature, data=iris_df)
plt.tight_layout()
plt.show()
```

```python
iris = sns.load_dataset("iris")
```

```python
iris.hist(edgecolor='black', linewidth=1.2, figsize=(12, 8))
plt.show()
```

```python
sns.pairplot(iris, hue="species", markers=["o", "s", "D"], diag_kind='hist')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
for i, feature in enumerate(iris.columns[:-1]):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(x="species", y=feature, data=iris)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
for i, feature in enumerate(iris.columns[:-1]):
    plt.subplot(2, 2, i + 1)
    sns.violinplot(x="species", y=feature, data=iris)
plt.tight_layout()
plt.show()
```
